```python
import pandas as pd
import numpy as np
```

```python
body_df = pd.read_csv("bodyPerformance.csv")
body_df.head()
```

```python
def map_gender(gender):
  if gender=='F':
    return 1
  else:
    return 0

body_df["gender"] = body_df["gender"].apply(lambda gender: map_gender(gender))
body_df.head()
```

```python
def map_body_class(bclass):
  if bclass=='A':
    return 1
  elif bclass=='B':
    return 2
  elif bclass=='C':
    return 3
  else:
    return 4
        

body_df["class"] = body_df["class"].apply(lambda bclass: map_body_class(bclass))
body_df.head()
```

```python
body_df.to_csv('bodyPerformance_2.csv', index=False)
```
